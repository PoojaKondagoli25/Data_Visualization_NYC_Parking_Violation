## ViolationTime vs Summons.csv

```python
def convert_time(time_str):
    if pd.isnull(time_str) or len(time_str) != 5:  # Check if time string is in expected format
        return None
    
    hour = int(time_str[:2])
    
    # Adjust hour for PM time
    if time_str[-1] == 'P' and hour != 12:
        hour += 12
    
    # Adjust hour for 12AM (midnight)
    if time_str[-1] == 'A' and hour == 12:
        hour = 0
    
    return hour

# Apply the function to each value in the 'Violation Time' column
df1['Violation Hour'] = df1['Violation Time'].map(convert_time).astype('Int64')  # Convert to Int64

df1.head(5)
```

```python
#merged_df.groupby('Violation category')['Issue Date'].count().reset_index()
all_counties = df1.groupby(['County Name','Street Name'])['Summons Number'].count().reset_index()

# Define a function to filter top 5 rows for each group
def top_5_desc(df):
    return df.nlargest(5, 'Summons Number')

# Apply the function to each group within 'County Name'
top_5_per_county = all_counties.groupby('County Name', group_keys=False).apply(top_5_desc)

# Reset index after filtering
top_5_per_county.reset_index(drop=True, inplace=True)
top_5_per_county.to_csv('Total five violating streets per county.csv', index=False)
print(top_5_per_county)
```
